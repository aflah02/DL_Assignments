```python
import numpy as np
from toolkit import *
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
x_train, y_train, x_val, y_val, x_test, y_test = load_data()
x_train.shape, y_train.shape, x_val.shape, y_val.shape, x_test.shape, y_test.shape
```

```python
model = NeuralNetwork(784, (28,28), [128], 10, relu, relu_prime, 'normal_init', mse, mse_prime, 'momentum', {'gamma': 0.9})
model.fit(x_train, y_train, x_val, y_val, epochs=10, learning_rate=0.01, batch_size=128, early_stopping=True, patience=5, verbose=True)
print(f"Accuracy: {model.score(x_test, y_test)}")
save_model(model, 'model_momentum')
```

```python
model = NeuralNetwork(784, (28,28), [128], 10, relu, relu_prime, 'normal_init', mse, mse_prime, 'nestrov_accelerated_gradient', {'gamma': 0.9})
model.fit(x_train, y_train, x_val, y_val, epochs=10, learning_rate=0.01, batch_size=128, early_stopping=True, patience=5, verbose=True)
print(f"Accuracy: {model.score(x_test, y_test)}")
save_model(model, 'model_nestrov_accelerated_gradient')
```

```python
model = NeuralNetwork(784, (28,28), [128], 10, relu, relu_prime, 'normal_init', mse, mse_prime, 'adagrad', {'epsilon': 1e-8})
model.fit(x_train, y_train, x_val, y_val, epochs=10, learning_rate=0.01, batch_size=128, early_stopping=True, patience=5, verbose=True)
print(f"Accuracy: {model.score(x_test, y_test)}")
save_model(model, 'model_adagrad')
```

```python
model = NeuralNetwork(784, (28,28), [128], 10, relu, relu_prime, 'normal_init', mse, mse_prime, 'rmsprop', {'epsilon': 1e-8, 'beta': 0.9})
model.fit(x_train, y_train, x_val, y_val, epochs=10, learning_rate=0.0001, batch_size=128, early_stopping=True, patience=5, verbose=True)
print(f"Accuracy: {model.score(x_test, y_test)}")
save_model(model, 'model_rmsprop')
```

```python
model = NeuralNetwork(784, (28,28), [128], 10, relu, relu_prime, 'normal_init', mse, mse_prime, 'adam', {'beta1': 0.9, 'beta2': 0.999, 'epsilon': 1e-6})
model.fit(x_train, y_train, x_val, y_val, epochs=10, learning_rate=0.0001, batch_size=128, early_stopping=True, patience=5, verbose=True)
print(f"Accuracy: {model.score(x_test, y_test)}")
save_model(model, 'model_adam')
```

```python
# Load the models
model_momentum = load_model('model_momentum')
model_nestrov_accelerated_gradient = load_model('model_nestrov_accelerated_gradient')
model_adagrad = load_model('model_adagrad')
model_rmsprop = load_model('model_rmsprop')
model_adam = load_model('model_adam')
models = [model_momentum, model_nestrov_accelerated_gradient, model_adagrad, model_rmsprop, model_adam]
```

```python
# Training Loss and Validation Loss V/s Epochs for different optimizers
for model in models:
    plt.plot(model.train_loss, label='Training Loss')
    plt.plot(model.val_loss, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(f"Training Loss and Validation Loss V/s Epochs for {model.optimizer} optimizer")
    plt.legend()
    plt.show()
```

```python
# Training Accuracy, Validation Accuracy V/s Epochs for different optimizers
for model in models:
    plt.plot(model.train_acc, label='Training Accuracy')
    plt.plot(model.val_acc, label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title(f"Training Accuracy and Validation Accuracy V/s Epochs for {model.optimizer} optimizer")
    plt.legend()
    plt.show()
```
